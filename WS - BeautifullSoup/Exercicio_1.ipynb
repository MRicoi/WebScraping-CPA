## NOSSO CODIGO

```python
# importar as bibliotecas

import requests
from bs4 import BeautifulSoup
import json
import re
```

```python
# pegar os 5000 links da wikipedia
links = []
i = 0
url = "https://pt.wikipedia.org/wiki/Dinheiro"
while len(links) < 5000:
    response = requests.get(url)
    soup = BeautifulSoup(response.content)
    link_total = soup.find_all('a')
    for link in link_total:
        if "href" in link.attrs.keys() and link['href'].startswith('/wiki/') and ":" not in link['href']:
            if link['href'] not in links and len(links) < 5000:
                links.append(link['href'])
    titulo = soup.select(".mw-page-title-main, .mw-first-heading")
    print(titulo[0].text)
    with open(f"Paginas HTML/{titulo[0].text}.html", "w", encoding="utf-8") as f:
        conteudo = response.content.decode("utf-8")
        f.write(conteudo)
    url = "https://pt.wikipedia.org" + links[i+1]
    i += 1
    print(len(links))
    print(url) 
```

```python
i = 0
for link in links:
    try:
        response = requests.get("https://pt.wikipedia.org" + link)
        soup = BeautifulSoup(response.content)
        infobox = soup.find_all("table", class_="infobox")

        if infobox:
            info_box_extraido = {}
            for tag in infobox[0].find_all('tr'):
                if tag.find(attrs={'scope':'row'}):
                    td_tag = tag.find_all('td')

                    if len(td_tag) > 1:
                        if td_tag[1].find('ul') or td_tag[1].find('div'):
                            valores = [filho.get_text(strip=True) for filho in td_tag[1].find_all("li")]
                        else:
                            valores = td_tag[1].get_text(strip=True)

                        info_box_extraido[td_tag[0].get_text(strip=True)] = valores

            titulo_info_tag = infobox[0].find("th")
            if titulo_info_tag:
                titulo_info = titulo_info_tag.get_text(strip=True)
            else:
                titulo_info = "Infobox_Sem_Título"

            titulo_info = re.sub(r'[<>:"/\\|?*]', '_', titulo_info)
            titulo_info = titulo_info[:100]

            with open(f"Paginas JSON/{titulo_info}.json", "w", encoding="utf-8") as json_file:
                json.dump(info_box_extraido, json_file, ensure_ascii=False, indent=4)

            print(f"Arquivo {titulo_info}.json salvo")
            i += 1
            print(i)

    except Exception as e:
        print(f"Erro com o link {link}: {e}")
        continue
```
